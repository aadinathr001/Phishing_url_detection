```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, Dense, Flatten, Dropout
from keras.utils import to_categorical
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense, Dropout, GlobalMaxPooling1D, Reshape
import os
from tensorflow.keras.models import load_model

from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk

# Define epochs for all models
epochs = 50
batchsize = 64
max_length = 300  # Adjust based on your dataset

# Load data
train = pd.read_csv(r"C:\Users\amrut\OneDrive\Desktop\trainvaltest\New folder (2)\train_data1.csv")
test = pd.read_csv(r"C:\Users\amrut\OneDrive\Desktop\trainvaltest\test_data.csv")
val = pd.read_csv(r"C:\Users\amrut\OneDrive\Desktop\trainvaltest\val_data.csv")

# Preprocess URLs into padded sequences
tokenizer = Tokenizer(char_level=True)
tokenizer.fit_on_texts(train['URL'])

max_length = 250  # Adjust based on your dataset

X_train_seq = tokenizer.texts_to_sequences(train['URL'])
X_test_seq = tokenizer.texts_to_sequences(test['URL'])
X_val_seq = tokenizer.texts_to_sequences(val['URL'])

X_train_padded = pad_sequences(X_train_seq, maxlen=max_length, padding='post')
X_test_padded = pad_sequences(X_test_seq, maxlen=max_length, padding='post')
X_val_padded = pad_sequences(X_val_seq, maxlen=max_length, padding='post')

vocab_size = len(tokenizer.word_index) + 1

# Labels
y_train = train['Label'].values
y_test = test['Label'].values
y_val = val['Label'].values

# Data distribution visualization
plt.figure(figsize=(3, 5))
train['Label'].value_counts().plot(kind='bar')
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.show()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
y_val = le.transform(y_val)


from keras.models import Sequential
from keras.layers import Embedding, Conv1D, MaxPooling1D, LSTM, Dense, Dropout, GlobalMaxPooling1D, Reshape
import os
from tensorflow.keras.models import load_model

model_path = r"C:\Users\amrut\anaconda scripts\cnn+lstm\val\model\newlr00011.keras"

if os.path.exists(model_path):
    print("Loading saved model...")
    model4 = load_model(model_path)
else:
    print("Training a new model...")
    
    # Define the model
    model4 = Sequential()

    # Embedding layer
    model4.add(Embedding(vocab_size, 64, input_length=max_length))

    # CNN layers
    model4.add(Conv1D(64, 3, activation='relu'))
    model4.add(MaxPooling1D(2))
    model4.add(Conv1D(128, 3, activation='relu'))
    model4.add(MaxPooling1D(2))

    # GlobalMaxPooling1D
    model4.add(GlobalMaxPooling1D())

    # Reshape for LSTM
    model4.add(Reshape((1, 128)))

    # LSTM layer
    model4.add(LSTM(128, return_sequences=False))

    # Dense layers
    model4.add(Dense(128, activation='relu'))
    model4.add(Dropout(0.2))
    model4.add(Dense(1, activation='sigmoid'))

    # Compile the model
    from tensorflow.keras.optimizers import Adam

    learning_rate = 0.00011  # Set your desired learning rate
    optimizer = Adam(learning_rate=learning_rate)
    model4.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy'])

    # Early stopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    # Train the model
    history = model4.fit(X_train_padded, y_train, batch_size=batchsize, epochs=epochs, 
                         validation_data=(X_test_padded, y_test), callbacks=[early_stopping])
    
    # Save the trained model
    model4.save(model_path)
    print(f"Model saved at {model_path}")
```

```python
import pickle
with open("training_history_newlr00011.pkl", "wb") as f:
    pickle.dump(history.history, f)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.figure(figsize=(10, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Model Accuracy')

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Model Loss')

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score

predictions = (model4.predict(X_test_padded) > 0.5).astype(int)

# Assuming you have true labels (y_test) and predicted labels (predictions)
tn, fp, fn, tp = confusion_matrix(y_test, predictions).ravel()

# Calculate metrics
specificity = tn / (tn + fp)
precision = precision_score(y_test, predictions)
recall = recall_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
accuracy = accuracy_score(y_test, predictions)

# Print results
print("Specificity:", specificity)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
print("Accuracy:", accuracy)
```
